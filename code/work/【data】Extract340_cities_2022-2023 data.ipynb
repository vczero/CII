```python
import pandas as pd
import os
import time
```

```python
PATH = '../微信公众号城市发布/data_ final_全国地级市_县发布/'
fls = [i for i in os.listdir(PATH) if '.csv' in i]
study_citys = pd.read_csv('./data/sa340.csv')
study_citys_name = list(set(study_citys['name']))
study_citys
```

```python
"""
Data from 2022-10-07 to 2023-10-06
"""
dfs = []
for i in fls:
    df = pd.read_csv(PATH + i, on_bad_lines='skip')
    dfs.append(df)
result = pd.concat(dfs, axis=0, ignore_index=True)
result['date'] = pd.to_datetime(result['time'])
filtered_df = result.query('date >="2022-10-07" and date < "2023-10-07" ')

"""
Data were selected according to the study area: 340 cities
"""
new_df = filtered_df[filtered_df['city_name'].isin(study_citys_name)]
new_df
```

```python
new_df.to_csv('./data/20221007_20231006_all.csv', index=None)
```
